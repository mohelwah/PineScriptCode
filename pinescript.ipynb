```python
# Iterate over each row in the DataFrame
for i, row in df.iterrows():
    high = row['high']
    low = row['low']

    # Check if current high is the highest in the last zigzag_len periods
    to_up = high >= talib.MAX(df['high'].values, timeperiod=zigzag_len)[i]
    to_down = low <= talib.MIN(df['low'].values, timeperiod=zigzag_len)[i]

    # Determine trend
    if i == 0:
        trend = 1
    else:
        trend = 1 if to_down else -1

    
    # Calculate last trend up since and low value/index
    last_trend_up_since = (to_up[:-1] != to_up[1:i]).argmax()
    low_val = talib.MIN(df['low'].values, timeperiod=max(last_trend_up_since, 1))[i]
    low_index = i - talib.BARSSINCE(df['low'].values, timeperiod=last_trend_up_since)[i]


    # Calculate last trend down since and high value/index
    last_trend_down_since = (to_down[:-1] != to_down[1:i]).argmax()
    high_val = talib.MAX(df['high'].values, timeperiod=max(last_trend_down_since, 1))[i]
    high_index = i - talib.BARSSINCE(df['high'].values, high_val)[i]
    
    # Add high and low points and their indices to the corresponding arrays
    if i > 0 and trend != df.at[i-1, 'trend']:
        if trend == 1:
            low_points_arr.append(low_val)
            low_index_arr.append(low_index)
        elif trend == -1:
            high_points_arr.append(high_val)
            high_index_arr.append(high_index)

    df.at[i, 'trend'] = trend

# Define a function to get high points
def f_get_high(ind):
    return [high_points_arr[-1-ind], high_index_arr[-1-ind]]

# Define a function to get low points
def f_get_low(ind):
    return [low_points_arr[-1-ind], low_index_arr[-1-ind]]

# Calculate market
market = np.ones(len(df))
for i, row in df.iterrows():
    if i > 0:
        market[i] = market[i-1]
    if i > 1:
        [h0, h0i] = f_get_high(0)
        [h1, h1i] = f_get_high(1)
        [l0, l0i] = f_get_low(0)
        [l1, l1i] = f_get_low(1)
        if l0 < l1 and l0 < l1 - abs(h0 - l1) * fib_factor:
            market[i] = -1
        elif h0 > h1 and h0 > h1 + abs(h0 - l1) * fib_factor:
            market[i] = 1

        df.at[i, 'market'] = market[i]


# Add market to dataframe
df['market'] = market

# Display output in a DataFrame
print(df)
    
```
